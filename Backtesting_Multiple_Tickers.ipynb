**RSI Backtesting**

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)

import backtrader as bt
import yfinance as yf
import pandas as pd
from google.colab import files

class RSIStrategy(bt.Strategy):
    params = {
        'rsi_period': 14,
        'rsi_overbought': 70,
        'rsi_oversold': 30,
    }

    def __init__(self):
        self.rsi = bt.indicators.RelativeStrengthIndex(period=self.params.rsi_period)
        self.order = None
        self.buyprice = None

    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        total_value = self.broker.getvalue()
        shares = self.get_order_position()
        print('%s, %s, Shares: %d, Total Value: %.2f' % (dt.isoformat(), txt, shares, total_value))

        # Extract transaction details from the log and append to the global list
        if 'BUY EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'BUY',
                'Price': price,
                'Total Value': total_value,
            })
        elif 'SELL EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'SELL',
                'Price': price,
                'Total Value': total_value,
            })

    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('BUY EXECUTED, Price: %.2f' % (order.executed.price))
                self.buyprice = order.executed.price
            else:
                self.log('SELL EXECUTED, Price: %.2f' % (order.executed.price))

            self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')

        self.order = None

    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('OPERATION PROFIT, GROSS %.2f, NET %.2f' % (trade.pnl, trade.pnlcomm))

    def next(self):
        self.log('Close, %.2f' % self.data.close[0])

        if self.order:
            return

        if not self.position:
            if self.rsi < self.params.rsi_oversold:
                cash = self.broker.get_cash()
                size = int(cash / self.data.close[0])
                self.order = self.buy(size=size)

        else:
            if self.rsi > self.params.rsi_overbought:
                self.sell(size=self.position.size)

    def get_order_position(self):
        if self.order:
            if self.order.isbuy():
                return self.order.executed.size
            elif self.order.issell():
                return -self.order.executed.size
        return 0

# Create an empty list to store transactions globally
transactions = []

def run_backtest(stock_symbol):
    cerebro = bt.Cerebro()
    cerebro.addstrategy(RSIStrategy)

    stock_data = yf.download(stock_symbol, period="1y")
    data = bt.feeds.PandasData(dataname=stock_data, name=stock_symbol)
    cerebro.adddata(data)

    cerebro.broker.set_cash(10000.0)

    starting_portfolio_value = cerebro.broker.getvalue()

    print(f'Starting Portfolio Value for {stock_symbol}: %.2f' % starting_portfolio_value)

    cerebro.run()

    # Plot the results
    cerebro.plot(style='candlestick', numfigs=1, iplot=False, volume=False)

    final_value = cerebro.broker.getvalue()
    final_cash = cerebro.broker.get_cash()
    pnl = final_value - starting_portfolio_value
    print(f'Final Portfolio Value for {stock_symbol}: {final_value:.2f}')
    print('Final Cash in Hand: %.2f' % final_cash)
    print(f'Final PnL for {stock_symbol}: {pnl:.2f}')

    return {
        'Symbol': stock_symbol,
        'Starting Portfolio Value': starting_portfolio_value,
        'Final Portfolio Value': final_value,
        'Final Cash in Hand': final_cash,
        'Final PnL': pnl,
    }

if __name__ == '__main__':
    stock_symbols = [
        "FL", "MRNA", "BABA", "PFE", "WBA", "AAL", "BNTX", "PYPL", "MMM", "DIS",
        "CVS", "VZ", "NKE", "TRIP", "UAL", "VOD", "T", "QCOM", "LEVI", "GM", "C",
        "LUV", "TSM", "GTLB", "BAC", "F", "SCHW", "GS", "WFC", "TGT"
    ]

    results = []

    for stock_symbol in stock_symbols:
        result = run_backtest(stock_symbol)
        results.append(result)

    # Create DataFrame from the transactions list
    transactions_df = pd.DataFrame(transactions)

    # Sort the DataFrame based on the "Date" column
    transactions_df = transactions_df.sort_values(by='Ticker')

    # Save the transactions_df to a CSV file
    transactions_df.to_csv("RSI_transactions.csv", index=False)

    print(transactions_df)

    # Convert the results to a DataFrame
    result_df = pd.DataFrame(results)

    # Sort the DataFrame based on the "Final PnL" column
    result_df = result_df.sort_values(by='Final PnL', ascending=False)

    # Save the result_df to a CSV file
    result_df.to_csv("RSI_results.csv", index=False)

    print(result_df)

    # Download the CSV files in Google Colab
    files.download("RSI_results.csv")
    files.download("RSI_transactions.csv")
```

**Bollinger Bands Backtesting**

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)

import backtrader as bt
import yfinance as yf
import pandas as pd
from google.colab import files

class BollingerBandStrategy(bt.Strategy):
    params = {
        'bollinger_period': 20,
        'bollinger_dev': 2,
    }

    def __init__(self):
        self.bollinger = bt.indicators.BollingerBands(self.data.close, period=self.params.bollinger_period, devfactor=self.params.bollinger_dev)
        self.order = None
        self.buyprice = None

    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        total_value = self.broker.getvalue()
        shares = self.get_order_position()
        print('%s, %s, Shares: %d, Total Value: %.2f' % (dt.isoformat(), txt, shares, total_value))

        # Extract transaction details from the log and append to the global list
        if 'BUY EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'BUY',
                'Price': price,
                'Total Value': total_value,
            })
        elif 'SELL EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'SELL',
                'Price': price,
                'Total Value': total_value,
            })

    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('BUY EXECUTED, Price: %.2f' % (order.executed.price))
                self.buyprice = order.executed.price
            else:
                self.log('SELL EXECUTED, Price: %.2f' % (order.executed.price))

            self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')

        self.order = None

    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('OPERATION PROFIT, GROSS %.2f, NET %.2f' % (trade.pnl, trade.pnlcomm))

    def next(self):
        self.log('Close, %.2f' % self.data.close[0])

        if self.order:
            return

        if not self.position:
            if self.data.close[0] < self.bollinger.lines.bot:
                # Calculate the order size based on available cash
                cash = self.broker.get_cash()
                size = int(cash / self.data.close[0])
                self.order = self.buy(size=size)

        else:
            if self.data.close[0] > self.bollinger.lines.top:
                # Sell all shares if close is above the upper Bollinger Band
                self.sell(size=self.position.size)

    def get_order_position(self):
        if self.order:
            if self.order.isbuy():
                return self.order.executed.size
            elif self.order.issell():
                return -self.order.executed.size
        return 0

# Create an empty list to store transactions globally
transactions = []

def run_backtest(stock_symbol):
    cerebro = bt.Cerebro()
    cerebro.addstrategy(BollingerBandStrategy)

    stock_data = yf.download(stock_symbol, period="1y")
    data = bt.feeds.PandasData(dataname=stock_data, name=stock_symbol)
    cerebro.adddata(data)

    cerebro.broker.set_cash(10000.0)

    starting_portfolio_value = cerebro.broker.getvalue()

    print(f'Starting Portfolio Value for {stock_symbol}: %.2f' % starting_portfolio_value)

    cerebro.run()

    final_value = cerebro.broker.getvalue()
    final_cash = cerebro.broker.get_cash()
    pnl = final_value - starting_portfolio_value
    print(f'Final Portfolio Value for {stock_symbol}: {final_value:.2f}')
    print('Final Cash in Hand: %.2f' % final_cash)
    print(f'Final PnL for {stock_symbol}: {pnl:.2f}')

    return {
        'Symbol': stock_symbol,
        'Starting Portfolio Value': starting_portfolio_value,
        'Final Portfolio Value': final_value,
        'Final Cash in Hand': final_cash,
        'Final PnL': pnl,
    }

if __name__ == '__main__':
    stock_symbols = [
        "FL", "MRNA", "BABA", "PFE", "WBA", "AAL", "BNTX", "PYPL", "MMM", "DIS",
        "CVS", "VZ", "NKE", "TRIP", "UAL", "VOD", "T", "QCOM", "LEVI", "GM", "C",
        "LUV", "TSM", "GTLB", "BAC", "F", "SCHW", "GS", "WFC", "TGT"
    ]

    results = []

    for stock_symbol in stock_symbols:
        result = run_backtest(stock_symbol)
        results.append(result)

    # Create DataFrame from the transactions list
    transactions_df = pd.DataFrame(transactions)

    # Sort the DataFrame based on the "Date" column
    transactions_df = transactions_df.sort_values(by='Ticker')

    # Save the transactions_df to a CSV file
    transactions_df.to_csv("bollinger_band_transactions.csv", index=False)

    print(transactions_df)

    # Convert the results to a DataFrame
    result_df = pd.DataFrame(results)

    # Sort the DataFrame based on the "Final PnL" column
    result_df = result_df.sort_values(by='Final PnL', ascending=False)

    # Save the result_df to a CSV file
    result_df.to_csv("bollinger_band_results.csv", index=False)

    # Download the CSV files in Google Colab
    files.download("bollinger_band_results.csv")
    files.download("bollinger_band_transactions.csv")
```

**Combination Between RSI and Bollinger Bands**

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)

import backtrader as bt
import yfinance as yf
import pandas as pd
from google.colab import files

class CombinedStrategy(bt.Strategy):
    params = {
        'rsi_period': 14,
        'rsi_overbought': 70,
        'rsi_oversold': 30,
        'bollinger_period': 20,
        'bollinger_dev': 2,
        'profit_target': 0.04,
        'stop_loss': 0.02,
    }

    def __init__(self):
        self.rsi = bt.indicators.RelativeStrengthIndex(period=self.params.rsi_period)
        self.bollinger = bt.indicators.BollingerBands(self.data.close, period=self.params.bollinger_period, devfactor=self.params.bollinger_dev)
        self.order = None
        self.buyprice = None

    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        total_value = self.broker.getvalue()
        shares = self.get_order_position()
        print('%s, %s, Shares: %d, Total Value: %.2f' % (dt.isoformat(), txt, shares, total_value))

        # Extract transaction details from the log and append to the global list
        if 'BUY EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'BUY',
                'Price': price,
                'Total Value': total_value,
            })
        elif 'SELL EXECUTED' in txt:
            price = self.data.close[0]
            transactions.append({
                'Date': dt,
                'Ticker': self.datas[0]._name,
                'Action': 'SELL',
                'Price': price,
                'Total Value': total_value,
            })

    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('BUY EXECUTED, Price: %.2f' % (order.executed.price))
                self.buyprice = order.executed.price
            else:
                self.log('SELL EXECUTED, Price: %.2f' % (order.executed.price))

            self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')

        self.order = None

    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('OPERATION PROFIT, GROSS %.2f, NET %.2f' % (trade.pnl, trade.pnlcomm))

    def next(self):
        self.log('Close, %.2f' % self.data.close[0])

        if self.order:
            return

        if not self.position:
            # Buy conditions: RSI oversold and close below lower Bollinger Band
            if self.rsi < self.params.rsi_oversold and self.data.close[0] < self.bollinger.lines.bot:
                cash = self.broker.get_cash()
                size = int(cash / self.data.close[0])
                self.order = self.buy(size=size)

        else:
            # Calculate PnL
            position_value_before = self.position.size * self.position.price
            position_value_after = self.position.size * self.data.close[0]
            pnl = position_value_after - position_value_before

            # Sell conditions: RSI overbought or close above upper Bollinger Band or profit/loss target reached
            if self.rsi > self.params.rsi_overbought or self.data.close[0] > self.bollinger.lines.top or pnl/position_value_before >= self.params.profit_target or pnl/position_value_before <= -self.params.stop_loss:
                self.sell(size=self.position.size)

    def get_order_position(self):
        if self.order:
            if self.order.isbuy():
                return self.order.executed.size
            elif self.order.issell():
                return -self.order.executed.size
        return 0

# Create an empty list to store transactions globally
transactions = []

def run_backtest(stock_symbol):
    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedStrategy)

    stock_data = yf.download(stock_symbol, period="1y")
    data = bt.feeds.PandasData(dataname=stock_data, name=stock_symbol)
    cerebro.adddata(data)

    cerebro.broker.set_cash(10000.0)

    starting_portfolio_value = cerebro.broker.getvalue()

    print(f'Starting Portfolio Value for {stock_symbol}: %.2f' % starting_portfolio_value)

    cerebro.run()

    final_value = cerebro.broker.getvalue()
    final_cash = cerebro.broker.get_cash()
    pnl = final_value - starting_portfolio_value
    print(f'Final Portfolio Value for {stock_symbol}: {final_value:.2f}')
    print('Final Cash in Hand: %.2f' % final_cash)
    print(f'Final PnL for {stock_symbol}: {pnl:.2f}')

    return {
        'Symbol': stock_symbol,
        'Starting Portfolio Value': starting_portfolio_value,
        'Final Portfolio Value': final_value,
        'Final Cash in Hand': final_cash,
        'Final PnL': pnl,
    }

if __name__ == '__main__':
    stock_symbols = [
        "FL", "MRNA", "BABA", "PFE", "WBA", "AAL", "BNTX", "PYPL", "MMM", "DIS",
        "CVS", "VZ", "NKE", "TRIP", "UAL", "VOD", "T", "QCOM", "LEVI", "GM", "C",
        "LUV", "TSM", "GTLB", "BAC", "F", "SCHW", "GS", "WFC", "TGT"
    ]

    results = []

    for stock_symbol in stock_symbols:
        result = run_backtest(stock_symbol)
        results.append(result)

    # Create DataFrame from the transactions list
    transactions_df = pd.DataFrame(transactions)

    # Sort the DataFrame based on the "Date" column
    transactions_df = transactions_df.sort_values(by='Ticker')

    # Save the transactions_df to a CSV file
    transactions_df.to_csv("combined_strategy_transactions.csv", index=False)

    print(transactions_df)

    # Convert the results to a DataFrame
    result_df = pd.DataFrame(results)

    # Sort the DataFrame based on the "Final PnL" column
    result_df = result_df.sort_values(by='Final PnL', ascending=False)

    # Save the result_df to a CSV file
    result_df.to_csv("combined_strategy_results.csv", index=False)

    # Download the CSV files in Google Colab
    files.download("combined_strategy_results.csv")
    files.download("combined_strategy_transactions.csv")
```
